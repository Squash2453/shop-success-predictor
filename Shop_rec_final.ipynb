```python
out_iqr = {}
for i in x:
    Q1=df[i].quantile(0.25)
    Q3=df[i].quantile(0.75)
    IQR=Q3-Q1
    Lower=Q1 - 1.5 * IQR
    Upper = Q3 +1.5 * IQR
    out=df[(df[i]<Lower) | (df[i]>Upper)]
    out_iqr[i]=out.shape[0]
out_iqr
```

```python
for i in x:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    Lower = Q1 - 1.5 * IQR
    Upper = Q3 + 1.5 * IQR
    df=df[(df[i]>Lower) & (df[i]<Upper)]
```

```python

sns.boxplot(df['Rent_Cost'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
scalar.fit(x_train)
x_train=scalar.transform(x_train)
x_test=scalar.transform(x_test)
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
score
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
y_pred1=svc.predict(x_test)
score1=accuracy_score(y_pred1,y_test)
score1
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(x_train,y_train)
y_pred2=gnb.predict(x_test)
score2=accuracy_score(y_pred2,y_test)
score2
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(criterion='entropy')
dtc.fit(x_train,y_train)
y_pred3=dtc.predict(x_test)
score3=accuracy_score(y_pred3,y_test)
score3
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=100,random_state=2)
rfc.fit(x_train,y_train)
y_pred4=rfc.predict(x_test)
score4=accuracy_score(y_pred4,y_test)
score4
```

```python
df2=pd.DataFrame({'Models':['KNN','SVC','GaussianNB','DecisionTree','RandomForest'],'Score':[score,score1,score2,score3,score4]})
df2
```

```python
plt.bar(df2['Models'],df2['Score'],color=['blue','green','red','cyan','magenta'])
plt.xlabel('Models')
plt.ylabel('Accuracy_score')
plt.title('Model Comparison')
```

```python

new_data = [[2, 1, 7, 1, 1, 1, 0, 2,
    15000, 680, 35000, 180, 120,
    0,1]]
probs = rfc.predict_proba(new_data)[0]

top3_idx = np.argsort(probs)[-3:][::-1]
classes = rfc.classes_
top3_classes = classes[top3_idx]
top3_probs = probs[top3_idx]

print("\nTop 3 Recommended Shops:\n")
for shop, p in zip(top3_classes, top3_probs):
    print(f"{shop}: {p*100:.2f}% success rate")
    
```
